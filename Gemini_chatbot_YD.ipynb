**Gemini Chat Basic**

```python
import google.generativeai as genai

genai.configure(api_key=API_KEY)
model = genai.GenerativeModel("gemini-pro")
response = model.generate_content("Write a sonnet about a President University student life that take Machine Learning class, but also make it rhyme.")
print(response.text)
```

**Gemini Chat w/ History**

```python
import google.generativeai as genai

genai.configure(api_key=API_KEY)
model = genai.GenerativeModel("gemini-pro")

# Removed generation_config from start_chat and applying it to send_message instead.
chatModel = model.start_chat(history=[])


# Example: Chat-like input in Colab
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Exiting the chat. Goodbye!")
        break
    else:
        # Applying generation_config to send_message method
        # The send_message method expects the prompt as a positional argument, not a keyword argument.
        result = chatModel.send_message(user_input, generation_config={"max_output_tokens": 200})
        print(f"AI: You said '{result.text}'")
```
