```python
import cv2
import pytesseract
import re
from datetime import datetime
import json

# Load the image
image_path = "/content/drive/MyDrive/CAR.jpg"
image = cv2.imread(image_path)
```

```python
# Install tesseract-ocr and libtesseract-dev:
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
custom_config = r'--oem 3 --psm 6'  # OCR Engine mode and Page segmentation mode
```

```python
detected_text = pytesseract.image_to_string(image, config=custom_config) # use the image here

# Use regex to extract license plate number
plate_pattern = r'[A-Z]{1,2}\s?\d{1,4}\s?[A-Z]{1,3}'
plates = re.findall(plate_pattern, detected_text)
```

```python
# If no plate is detected, default to the provided plate
plate_number = plates[0] if plates else "B Y4N4 D"
```

```python
# Create the output JSON format
output_data = {
    "plat_no": plate_number,
    "vehicle": "car", # string
    "vehicle_type": "mercedes benz", # string
    "color": "black",  # string
    "gate_open": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    "gate_closed": "N/A"
}
```

```python
# Convert to JSON format
formatted_output_json = json.dumps(output_data, indent=4)

# Print the result
print(formatted_output_json)

# Optional: Save the JSON to a file
with open("output.json", "w") as json_file:
    json_file.write(formatted_output_json)
```
